##### Thay thế bằng giá trị: mean, median, mode...

```python
data = {
'a': [10, 9.5, 5, np.nan, 8.1, 2, 8.3, np.nan, 6, 8],
'b': [1, 8, 3, np.nan, 6, 8, 2, np.nan, 4, 8],
'c': ['fwd', '4wd', np.nan, 'fwd', 'rwd', np.nan, 'fwd', 'rwd', '4wd', 'fwd']
}
df = pd.DataFrame(data)
df
```

```python
df['a'] = df['a'].astype(float).fillna(value=df['a'].mean())
df['b'] = df['b'].astype("Int64").fillna(value=df['b'].median())
df['c'] = df['c'].fillna(value=df['c'].mode()[0])
df
```

```python
data = {
    'a': [10, 9.5, 5, np.nan, 8.1, 2, 8.3, np.nan, 6, 8],
    'b': [1, 8, 3, np.nan, 6, 8, 2, np.nan, 4, 8],
    'c': ['fwd', '4wd', np.nan, 'fwd', 'rwd', np.nan, 'fwd', 'rwd', '4wd', 'fwd']
}

df = pd.DataFrame(data)

df.fillna(
    {
        'a': -1,
        'b': -1,
        'c': 'Missing'
    },
    inplace=True
)
df
```

##### Dung nhan du lieu bi thieu (Missingness Indicator)

```python
df['a_missing'] = df['a'].isna().astype('int')
df['b_missing'] = df['b'].isna().astype('int')
```
