 # If - Else

```python
def tro_choi(nguoi_ra, may_ra):
    if nguoi_ra == may_ra:
        return "Hoa"
    elif (nguoi_ra == 'keo' and may_ra == 'bua') or (nguoi_ra == 'bua' and may_ra == 'bao') or (nguoi_ra == 'bao' and may_ra == 'keo'):
        return "Nguoi thang"
    else:
        return "May thang"

nguoi_ra = input("Nguoi ra ")
may_ra = input("May ra: ")
ket_qua = tro_choi(nguoi_ra, may_ra)
ket_qua
```

# Function

### func

```python
def tinh_tong(n):
    if n < 0:
        return 0
    else:
        return n + tinh_tong(n-1)
print(tinh_tong(5))
```

```python
def tinh(a:int, b:float) -> tuple:
    cong = a + b
    tru = a - b
    nhan = a * b
    chia = a / b if b != 0 else None
    return (cong, tru, nhan, chia)
print(tinh(5, 2.5))
```

```python
def my_function(x, y=1, *z, **kw):
    a = x + y
    for i in z:
        a += i
    for k, v in kw.items():
        a += v
    return a

my_function(2,1,3,4,5, m=6, n=7)
```
