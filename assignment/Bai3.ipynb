```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y1': [10, 15, 13, 17, 14],
    'y2': [16, 5, 11, 9, 15]
})
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.subplot(2,1,1)
plt.plot(df['x'], df['y1'], color='r')
plt.title('Biểu đồ đường')
```

```python
plt.subplot(2,2,4)
plt.bar(df['x'], df['y2'], color='b')
plt.title('Biểu đồ cột')
```

```python
plt.show
```

```python
s = pd.Series([1, 2, 3, 3])
plot = s.plot(kind='bar', title='My Plot')
```

```python
df.plot(x='x', y=['y1','y2'], kind='line', marker='o')
plt.title('Biểu đồ đường từ DataFrame')
plt.xlabel('X-axis')
plt.ylabel('Y1-axis')
plt.grid(True)
plt.show()
```
