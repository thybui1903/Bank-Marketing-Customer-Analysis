##### Câu 1. Đánh giá sơ bộ trình trạng bị khuyết dữ liệu. Chú ý: Giá trị khuyết thực tế là “?”, “”, “”, “nan”, “NAN”

```python
list_missing_values = ['?', 'NAN', '.', 'nan', '-']
for col in df.columns:
    for miss in list_missing_values:
        miss_count = df[col][df[col] == miss].count()
        if miss_count > 0:
            print(f'column {col} has {miss_count} missing values represented by {miss}')      
```

#### Câu 2. Xử lý giá trị bị khuyết theo kịch bản sau:

- "bedrooms": thay thế bằng giá trị xuất hiện nhiều lần; 
- "bathrooms": thay thế bằng giá trị trung tâm; 
- "view": thay thế bằng giá trị xuất hiện nhiều lần; 
- "sqft_basement" thay thế bằng giá trị cố định; 
- "sqft_lot" thay thế bằng giá trị trung bình; 
- "floors" thay thế bằng giá trị cố định; 
- "yr_renovated" dùng nhãn dữ liệu bị khuyết;
- Nếu còn lại các biến bị khuyết thì loại bỏ mẫu dữ liệu


##### - "bedrooms": thay thế bằng giá trị xuất hiện nhiều lần; 

```python
df['bedrooms'].fillna(df['bedrooms'].mode()[0], inplace=True)
df['bedrooms']
df['bedrooms_mode'] = df['bedrooms']
```

##### - "bathrooms": thay thế bằng giá trị trung tâm; 

```python
df['bathrooms'].replace('?', np.nan, inplace=True)
df['bathrooms'] = pd.to_numeric(df['bathrooms'], errors='coerce')
df['bathrooms'].fillna(df['bathrooms'].median(), inplace=True)
df['bathrooms'].isnull().sum()
df['bathrooms_median'] = df['bathrooms']
```

##### - "view": thay thế bằng giá trị xuất hiện nhiều lần; 

```python
df['view'][df['view'].isnull()==True]
```

```python
df['view'].fillna(df['view'].mode()[0], inplace=True)
df['view'].isnull().sum()
df['view_mode'] = df['view']
```

##### - "sqft_basement" thay thế bằng giá trị cố định; 

```python
df['sqft_basement'].replace('NAN', np.nan, inplace=True)
df['sqft_basement'][df['sqft_basement'].isnull()==True]
```

```python
df['sqft_basement'].fillna(0, inplace=True)
df['sqft_basement'].isnull().sum()
df['sqft_basement_fill0'] = df['sqft_basement']
```

##### - "sqft_lot" thay thế bằng giá trị trung bình; 

```python
df['sqft_lot'].replace('NAN', np.nan, inplace=True)
df['sqft_lot'][df['sqft_lot'].isnull()==True]
df['sqft_lot'] = pd.to_numeric(df['sqft_lot'], errors='coerce')
df['sqft_lot'].fillna(df['sqft_lot'].mean(), inplace=True)
df['sqft_lot_mean'] = df['sqft_lot']
```

##### - "floors" thay thế bằng giá trị cố định;

```python
df['floors'][df['floors'].isnull()==True]
```

```python
df['floors'].replace('NAN', np.nan, inplace=True)
df['floors'].fillna(1, inplace=True)
df['floors'].isnull().sum()
df['floors_fill1'] = df['floors']
```

##### - "yr_renovated" dùng nhãn dữ liệu bị khuyết;

```python
df['yr_renovated'].replace('NAN', np.nan, inplace=True)
df['yr_renovated'][df['yr_renovated'].isnull()==True]
```

```python
df['yr_renovated_missing'] = np.where(df['yr_renovated'].isnull(), 1, 0)
df['yr_renovated_missing'].value_counts()
```

##### - Nếu còn lại các biến bị khuyết thì loại bỏ mẫu dữ liệu

```python
df['price'].replace('?', np.nan, inplace=True)

list_cols = ['bedrooms', 'bathrooms', 'view', 'sqft_basement', 'sqft_lot', 'floors', 'yr_renovated', 'price']
# Lấy danh sách các cột ngoài list_cols
other_cols = [col for col in df.columns if col not in list_cols]

# Drop các dòng còn NaN ở các cột ngoài list_cols
df.dropna(subset=other_cols, inplace=True)

# Reset lại index
df.reset_index(drop=True, inplace=True)
```

Câu 4. Chọn các cột có giá trị phân loại chuẩn hóa theo:
- MinMaxScaler
- MaxAbsScaler
- RobustScaler
- StandardScaler

```python
df
```

```python
# MinMaxScaler
MinMaxScaler_cols = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

for col in MinMaxScaler_cols:
    df[col + '_MinMaxScaler'] = scaler.fit_transform(df[[col]])
```

```python
# MaxAbsScaler
from sklearn.preprocessing import MaxAbsScaler
MaxAbsScaler_cols = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']
scaler = MaxAbsScaler()
for col in MaxAbsScaler_cols:
    df[col + '_MaxAbsScaler'] = scaler.fit_transform(df[[col]])
```

```python
# RobustScaler
from sklearn.preprocessing import RobustScaler
RobustScaler_cols = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']
scaler = RobustScaler()
for col in RobustScaler_cols:
    df[col + '_RobustScaler'] = scaler.fit_transform(df[[col]])
```

```python
# StandardScaler
from sklearn.preprocessing import StandardScaler
StandardScaler_cols = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']
scaler = StandardScaler()
for col in StandardScaler_cols:
    df[col + '_StandardScaler'] = scaler.fit_transform(df[[col]])
```

#### Câu 5.
Mã hóa dữ liệu từ dạng phân loại sang dạng số theo các phương pháp sau:
- Label encoding
- Ordinal Encoding
- One-Hot encoding

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categorical_cols = ['view']
for col in categorical_cols:
    df[col + '_LabelEncoder'] = le.fit_transform(df[col])
```

```python
# Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder    
oe = OrdinalEncoder()
ordinal_cols = ['view', 'condition', 'grade_v2']
df[ordinal_cols] = oe.fit_transform(df[ordinal_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first')

encoded = ohe.fit_transform(df[['waterfront']])

encoded_cols = ohe.get_feature_names_out(['waterfront'])

encoded_df = pd.DataFrame(encoded, columns=encoded_cols, index=df.index)

df = pd.concat([df, encoded_df], axis=1)
```
